```python
import random

class NakliLLM:

  def __init__(self):
    print('LLM created')

  def invoke(self, prompt):

    response_list = [
        'Delhi is the capital of India',
        'IPL is a cricket league',
        'AI stands for Artificial Intelligence'
    ]

    return {'response': random.choice(response_list)}
```

```python
class template:
  def __init__(self,temp,input):
    self.temp = temp
    self.input = input

  def invoke(self,dicto):
    return self.temp.format(**dicto)


```

```python
class runnable:
  def __init__(self,lis):
    self.lis=lis

  def invoke(self,dicto):
    for i in self.lis:
      dicto = i.invoke(dicto)
      print(dicto)

    return dicto
```

```python
templat=template('The capital of {country} is {capital}',{'country','capital'})
llm=NakliLLM()
```

```python
run=runnable([templat,llm])
print(
run.invoke({'country':'india','capital':'delhi'}))
```
